```python
import numpy as np
import pandas as pd
import string
import re
```

```python
message = pd.read_csv(r'C:\Users\Akhtar\Downloads\Messages.csv')
```

```python
keyword = pd.read_csv(r'C:\Users\Akhtar\Downloads\keywords.csv')
```

```python
message.columns = ['view']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
corpus = []
for i in range (0, 5878):
    view = re.sub('[^a-zA-Z]', ' ', message['view'][i])
    view = view.lower()
    view = view.split()
    ps = PorterStemmer()
    view = [ps.stem(word) for word in view if not word in set(stopwords.words('english'))]
    view = ' '.join(view)
    corpus.append(view)
```
